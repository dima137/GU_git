```python
import numpy as np
import pandas as pd
from matplotlib import pyplot
print(np.__version__)
#from sklearn.metrics import ConfusionMatrixDisplay
#import seaborn
```

```python
ML_folder = '../../data_in/ML_catalogs/'
fn_ML = ML_folder + '4FGL-DR4_v34_cov_shift_vs_prior_shift_prediction.csv'
list_fn = '../../data_in/GU_lists/list_bright_GUs.csv'
fn_out = '../../data_in/GU_lists/list_bright_GUs_ML_classes.csv'
```

```python
ML_df = pd.read_csv(fn_ML, index_col=0)
GU_df = pd.read_csv(list_fn, index_col=0)
```

```python
names = [name for name in GU_df.index if name in ML_df.index]
names_na = [name for name in GU_df.index if name not in names]
```

```python
add_cols = ['RF_class', 'PDF_class', 'Gauss_PDF']
for col in add_cols:
    GU_df.loc[names, col] = ML_df.loc[names, col]
    if col.find('class') > -1:
        GU_df.loc[names_na, col] = 'na'
    else:
        GU_df.loc[names_na, col] = 0.
        
```
